```python
#Take a paragraph as Input and output the top three most repeated words

# Input paragraph
paragraph = "This is a sample paragraph. This paragraph contains some sample words for testing. Sample words are repeated to test this word count exercise."
paragraph_lower = paragraph.lower()
words = paragraph_lower.split()
# print(words)

# create a dictionary to store word count

word_count = {}

# count the occurance of each word

for word in words:
	word_count[word] = word_count.get(word, 0) + 1
# print(word_count)

sorted_word_count = list(word_count.items())
# print(sorted_word_count)
# print(len(sorted_word_count))

# Custom sorting without using a function
#same logic as bubble sort
for i in range(len(sorted_word_count)):
    for j in range(i + 1, len(sorted_word_count)):
        if sorted_word_count[i][1] < sorted_word_count[j][1]:
            sorted_word_count[i], sorted_word_count[j] = sorted_word_count[j], sorted_word_count[i]

# print(sorted_word_count)

# Print the top three most repeated words
for i in range(min(3, len(sorted_word_count))):
    print(sorted_word_count[i][0])
```

```python
# You are given a string sentence that consist of words separated by spaces. Each word consists of lowercase and uppercase letters only.

# We would like to convert the sentence to "Goat Latin" (a made-up language similar to Pig Latin.) The rules of Goat Latin are as follows:

# If a word begins with a vowel ('a', 'e', 'i', 'o', or 'u'), append "ma" to the end of the word.
# For example, the word "apple" becomes "applema".

# If a word begins with a consonant (i.e., not a vowel), remove the first letter and append it to the end, then add "ma".
# For example, the word "goat" becomes "oatgma".

# Add one letter 'a' to the end of each word per its word index in the sentence, starting with 1.
# For example, the first word gets "a" added to the end, the second word gets "aa" added to the end, and so on.

# Return the final sentence representing the conversion from sentence to Goat Latin.

 

# Example 1:

# Input: sentence = "I speak Goat Latin"
# Output: "Imaa peaksmaaa oatGmaaaa atinLmaaaaa"
# Example 2:

# Input: sentence = "The quick brown fox jumped over the lazy dog"
# Output: "heTmaa uickqmaaa rownbmaaaa oxfmaaaaa umpedjmaaaaaa overmaaaaaaa hetmaaaaaaaa azylmaaaaaaaaa ogdmaaaaaaaaaa"

sentence = "I speak Goat Latin"
vowels = 'aeiouAEIOU'
words = sentence.split()
print(words)

goat_latin_sentence = ""

for i in range(len(words)):
	word = words[i]
	#print(word)
	#print(word[1:])
	if word[0] in vowels:
		goat_latin_word= word + "ma"
	else:
		goat_latin_word = word[1:] + word[0] + "ma"
	goat_latin_word = goat_latin_word + "a" * (i + 1)
	print(goat_latin_word)
	goat_latin_sentence = goat_latin_sentence + goat_latin_word

	goat_latin_sentence = goat_latin_sentence + " "


goat_latin_sentence = goat_latin_sentence.strip()

print(goat_latin_sentence)
```

```python

# Input words and k
words = ["the","day","is","sunny","the","the","the","sunny","is","is"]
k = 4


# Create a dictionary to count the frequency of each word
freq_dict = {}

for word in words:
	if word in freq_dict:
		freq_dict[word] = freq_dict[word] + 1
	else:
		freq_dict[word] = 1

print(freq_dict)

# Convert the dictionary to a list of tuples (word, frequency)

freq_list=[]

for word in freq_dict:
	freq_list.append((word, freq_dict[word]))

print(freq_list)

# Step 3: Sort the list by frequency (descending) and then by word (lexicographically)
for i in range(len(freq_list)):
    for j in range(i + 1, len(freq_list)):
        # Compare frequencies
        if freq_list[i][1] < freq_list[j][1]:
            freq_list[i], freq_list[j] = freq_list[j], freq_list[i]
        # If frequencies are the same, compare words lexicographically
        elif freq_list[i][1] == freq_list[j][1] and freq_list[i][0] > freq_list[j][0]:
            freq_list[i], freq_list[j] = freq_list[j], freq_list[i]

# Step 4: Extract the top k words
top_k_words = []
for i in range(k):
    top_k_words.append(freq_list[i][0])

# Print the result
print(top_k_words)  # Output: ["i", "love"]
```
