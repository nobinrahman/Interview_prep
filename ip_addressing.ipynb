```python
#Write code to generate hosts addresses for a given network add and mask.
#Assume there is a lib function that converts ip to num and num to ip

import ipaddress

# Define the network
network = ipaddress.IPv4Network('192.168.10.0/24')
print(network)

# Get the first and last host addresses
first_host = network.network_address + 1
last_host = network.broadcast_address - 1

# Print the range of host addresses
print(f"The range of host addresses for network {network.network_address}/{network.prefixlen} is:")
print(f"First Host: {first_host}")
print(f"Last Host: {last_host}")

ip_list = []

for ip in range(int(first_host), int(last_host) + 1):
	ip_addr = ipaddress.IPv4Address(ip)
	ip_list.append(str(ip_addr))

#print(ip_list)

### Find host address
network = ipaddress.IPv4Network('192.168.1.0/24')
for ip in network.hosts():
    print(ip)
```

```python
### Find address type 

import ipaddress

def ip_type(ip_addr):
    try:
        addr = ipaddress.ip_address(ip_addr)
        if addr.is_private:
            return "Private"
        elif addr.is_global:
            return "Public"
        elif addr.is_multicast:
            return "Multicast"
        elif addr.is_loopback:
            return "Loopback"
        else:
            return "Unknown"
    except ValueError:
        return "Invalid IP Address"

# Example usage
ip1 = '192.168.1.1'
ip2 = '8.8.8.8'
ip3 = '224.0.0.1'
ip4 = '127.0.0.1'
ip5 = '256.1.1.1'

print(f"{ip1}: {ip_type(ip1)}")
print(f"{ip2}: {ip_type(ip2)}")
print(f"{ip3}: {ip_type(ip3)}")
print(f"{ip4}: {ip_type(ip4)}")
print(f"{ip5}: {ip_type(ip5)}")
```

```python
# Write a list of IP addresses starting from a certain subnet to certain subnet
import ipaddress

def generate_ip_list(start_subnet, end_subnet):
    start_ip = ipaddress.IPv4Address(start_subnet.split('/')[0])
    end_ip = ipaddress.IPv4Address(end_subnet.split('/')[0])

    ip_list = []
    for ip_int in range(int(start_ip), int(end_ip) + 1):
        ip_addr = ipaddress.IPv4Address(ip_int)
        ip_list.append(str(ip_addr))

    return ip_list

# Example usage
start_subnet = '192.168.0.0/24'
end_subnet = '192.168.1.0/24'
ip_list = generate_ip_list(start_subnet, end_subnet)

print(ip_list)
```
